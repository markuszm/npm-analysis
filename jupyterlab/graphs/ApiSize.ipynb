```python
import pandas as pd
import json

import numpy as np
import matplotlib
import matplotlib.pyplot as plt

fileName = 'apiSizeWithHeuristic.json'

usage = pd.read_json(fileName, orient='records')

Y = usage['Size']
```

```python
plt.figure(figsize=(8,6), dpi=100)
n, bins, patches = plt.hist(Y, bins='doane')
plt.grid(axis='y', alpha=0.75)
# plt.gca().set_xticklabels(['{:.0f}%'.format(x*100) for x in plt.gca().get_xticks()]) 
plt.xlabel('Size of API')
plt.ylabel('Packages')
print(n,bins)
plt.savefig('{}.png'.format(fileName))
```

```python
bars = [
    usage[usage['Size'] == 1],
    usage[usage['Size'] == 2],
    usage[usage['Size'] == 3],
    usage[usage['Size'] == 4],
    usage[usage['Size'] == 5],
    usage[usage['Size'].between(6, 9)], 
    usage[usage['Size'].between(10, 99)],
    usage[usage['Size'].between(100, 999)],
    usage[usage['Size'].between(1000, 9999)]    
]

Y = []

for b in bars:
    Y.append(len(b))
X = np.arange(0, len(Y), step=1)

maxSize = usage['Size'].max()

matplotlib.rcParams.update({'font.size': 14})
plt.figure(figsize=(10,9), dpi=90)
plt.bar(X, Y, color="blue")

# for x,y in zip(X,Y):
#     plt.text(x, y, y, ha='center', va= 'bottom')


plt.xlabel('Size of API')
plt.ylabel('Packages')
# plt.ylim(0, 700000)
plt.xticks(np.arange(0, len(Y), step=1), ['1', '2',  '3', '4', '5', '6-9', '10-99', '100-999', '1000-{}'.format(maxSize)], rotation=-30)

plt.savefig('{}.png'.format(fileName))
```
