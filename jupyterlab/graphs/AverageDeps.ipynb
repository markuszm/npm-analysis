```python
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6), dpi=90)

plt.rcParams.update({'font.size': 14})



def transformDf(df, key):
    df[key] = pd.to_datetime(df[key])
    df = df.set_index(key)
    return df

dataAvg = pd.read_json('averagePackageCost.json')
averagePackageCost = transformDf(dataAvg, "Key")

plt.xlabel('Time')
plt.ylabel('Average Dependencies (log scale)')
plt.plot(averageDeps["averagesOut"], label='direct dependencies')
plt.plot(averagePackageCost["Value"], label='transitive dependencies')
# plt.scatter(averageDeps["averagesOut"],averagePackageCost["Value"])
plt.legend(loc='best', frameon=False)
plt.gca().set_yscale("log")
plt.gca().set_yticklabels(['{:.0f}'.format(x)
                           for x in plt.gca().get_yticks()])
plt.savefig("average_direct_vs_transitive_dependencies_evolution_log.png")
```
