```python
import pandas as pd
import json

import numpy as np
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams.update({'font.size': 14})

averageDeps = pd.read_json('averageDeps.json')
averageDeps
# averageDeps = pd.read_json('averageDepsWithDep.json')
```

```python
plt.figure(figsize=(8,6), dpi=90)
plt.rcParams.update({'font.size': 14})

maintainerCountGrowth = pd.read_json('maintainerCountGrowth.json')    

plt.xlabel('Time')
plt.ylabel('Count')
# plt.plot(averageDeps["packageCount"], maintainerCountGrowth)
plt.plot(averageDeps["packageCount"], label="Packages")
plt.plot(maintainerCountGrowth, label="Maintainers")
plt.legend(loc='best', frameon=False)
plt.savefig("package_and_maintainer_count_evolution.png")
```

```python
plt.figure(figsize=(8,6), dpi=90)
plt.rcParams.update({'font.size': 14})

plt.xlabel('Number of packages')
plt.ylabel('Number of maintainers')
# plt.plot(averageDeps["packageCount"], maintainerCountGrowth)
plt.plot(averageDeps["packageCount"], maintainerCountGrowth)
# plt.legend(loc='best', frameon=False)
plt.savefig("package_and_maintainer_linear_together.png")
```
