```python
import pandas as pd

nodeJsModules = set(pd.read_json('nodejsModules.json', typ='series'))

data = pd.read_json('usedDeps.json')
```

```python
cleaned = data[data.Result.apply(isinstance, args=(dict,))]

counts = {}
for n in nodeJsModules:
    counts[n] = 0

for res in cleaned.Result:
    req = res['Required']
    imp = res['Imported']
    if len(req) == 0 and len(imp) == 0:
        continue
    for r in req:
        if r in nodeJsModules:
            counts[r] = counts[r] + 1

    for i in imp:
        if i in nodeJsModules:
            counts[i] = counts[i] + 1


sorted_by_value = sorted(counts.items(), key=lambda kv: kv[1], reverse=True)
```

```python
import csv

with open('topNodejsModulesRequired.csv', mode='w') as file:
    writer = csv.writer(file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)

    for row in sorted_by_value:
        writer.writerow([row[0], row[1]])
```
